##### [실습] 교차 검증

```python
# import 구문
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
fish_df = pd.read_csv('./data/fish.csv')
fish_df['Species'].unique()

display(fish_df)

le = LabelEncoder()
fish_df['Species'] = le.fit_transform(fish_df['Species'])

fish_df
```

```python
from sklearn.model_selection import train_test_split
X = fish_df.drop('Species', axis=1)
y = fish_df['Species']

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

- 생선 다중 분류 with cross_val_score

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

lr_clf = LogisticRegression()

scores = cross_val_score(lr_clf, X_train, y_train, cv=5, scoring='accuracy')

print('훈련별 정확도:', scores)
print('모델 정확도:', np.mean(scores))
```

- 생선 다중 분류 with GridSearchCV

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV


xgb_clf = XGBClassifier()

params = {
    'n_estimators': range(60,101),
    'max_depth': range(3,10),
    'learning_rate': [0.01, 0.1]
}

grid = GridSearchCV(xgb_clf, params, scoring='accuracy', cv=5)
grid.fit(X_train, y_train)


print(f"최적의 파라미터:", grid.best_params_)
print(f"최적의 모델 객체:", grid.best_estimator_)
print(f"최적화된 점수:", grid.best_score_)
```
